```python
import numpy as np
x = np.arange(0,10)
y = x*2
z = x**2
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('seaborn-white')
```

```python
#Даалгавар 1
fig = plt.figure()
ax = plt.axes()
ax.plot(x,y)
```

```python
#Даалгавар 2
ax1 = plt.axes()
ax2 = plt.axes([0.25,0.5,0.2,0.2])
fig = plt.figure()
ax1.plot(x,y, color = 'brown')
ax2.plot(x,y)
```

```python
#Даалгавар 3, 4
fig = plt.figure()
ax3 = plt.subplot(2,1,1)
ax3.plot(x,y,linestyle = 'dashed')
ax4 = plt.subplot(2,1,2)
ax4.plot(x,z,color = 'red')
```
