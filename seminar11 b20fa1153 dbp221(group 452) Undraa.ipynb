```python
#Даалгавар1:Column chart:
import matplotlib.pyplot as plt
 
#data
x = ['Python', 'C', 'Java', 'C++', 'C#', 'VB', 'JS', 'As', 'SQL', 'swift',]
h = [12.9, 11.80, 10.12, 7.73, 6.40, 5.40, 2.30, 2.25, 1.79, 1.76,]
c = 'teal', 'red', 'orange', 'pink', 'black', 'blue', 'coral', 'green', 'purple'
 
#bar plot
plt.bar(x, height = h, color = c)
 
plt.show()
```

```python
#Даалгавар1:Column chart:
import matplotlib.pyplot as plt
 
#data
x = ['R', 'PHP', 'CVB', 'Groovy', 'Ruby', 'D', 'Fortran', 'Perl', 'Go', 'MATPLAB',]
h = [1.58, 1.50, 1.27, 1.23, 1.16, 1.14, 1.04, 0.96, 0.95, 0.92,]
c = 'brown', 'purple', 'orange', 'pink', 'black', 'blue', 'coral', 'green', 'red'
 
#bar plot
plt.bar(x, height = h, color = c)
 
plt.show()
```

```python
# Даалгавар1:Pie chart:
labels = 'Python', 'C', 'Java', 'C++'
sizes = [12.9, 11.80, 10.12, 7.73,]
explode = (0, 0.1, 0, 0)  # only "explode" the 2nd slice (i.e. 'Hogs')

fig1, ax1 = plt.subplots()
ax1.pie(sizes, explode=explode, labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.

plt.show()
```
