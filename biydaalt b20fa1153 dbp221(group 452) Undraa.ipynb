```python
import pandas as pd
import numpy as np
```

```python
irts1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Ирц")
irts = irts1.dropna(axis = 1, how = 'all')
irts
```

```python
irts1 = irts.replace('и', 1)
irts2 = irts1.replace('т', 0)
irsen = irts2.sum(axis = 1)
onoo_irts = irsen * 10 / 32
print(round(onoo_irts, 2))
```

```python
seminar1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Семинар")
seminar = seminar1.dropna(axis = 1, how = 'all')
seminar
```

```python
hiisen = seminar.sum(axis = 1)
onoo_sem = hiisen * 25 / 19
print(round(onoo_sem, 2))
```

```python
exam1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Шалгалт")
exam = exam1.dropna(axis = 1, how = 'all')
exam
```

```python
exam1 = exam.fillna(0)
onoo_exam = exam1.sum(axis = 1)
print(round(onoo_exam, 2))
```

```python
biydaalt1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Бие даалт")
biydaalt = biydaalt1.dropna(axis = 1, how = 'all')
biydaalt
```

```python
onoo_biydaalt = biydaalt.sum(axis = 1)
print(round(onoo_biydaalt, 2))
```

```python
humuujil1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Хүмүүжил хандлага")
humuujil = humuujil1.dropna(axis = 1, how = 'all')
humuujil
```

```python
onoo_humuujil = humuujil.sum(axis = 1)
print(round(onoo_humuujil, 2))
```

```python
bonus1 = pd.read_excel("Бие даалт - 1.xlsx", index_col=0, header=0, sheet_name="Нэмэлт оноо")
bonus = bonus1.dropna(axis = 1, how = 'all')
bonus
```

```python
onoo_bonus = bonus.sum(axis = 1)
print(round(onoo_bonus, 2))
```

```python
onoo_niit = onoo_irts+onoo_sem+onoo_exam+onoo_biydaalt+onoo_humuujil+onoo_bonus
```

```python
niit = pd.DataFrame({'Ирц': round(onoo_irts,2),
                     'Семинар': round(onoo_sem,2),
                     'Шалгалт': onoo_exam,
                     'Бие даалт': onoo_biydaalt,
                     'ХХ': onoo_humuujil,
                     'Нэмэлт': onoo_bonus,
                     'Нийт': round(onoo_niit,2)})
niit
```

```python
def unelgee(x):
    if x>=96:
        return "A"
    elif x>=90 and x<=95:
        return "A-"
    elif x>=87 and x<=89:
        return "B+"
    elif x>=83 and x<=86:
        return "B"
    elif x>=80 and x<=82:
        return "B-"
    elif x>=77 and x<=79:
        return "C+"
    elif x>=73 and x<=76:
        return "C"
    elif x>=70 and x<=72:
        return "C-"
    elif x>=67 and x<=69:
        return "D+"
    elif x>=63 and x<=66:
        return "D"
    elif x>=60 and x<=62:
        return "D-"
    else:
        return "F"
niit['Үнэлгээ'] = round(niit['Нийт']).apply(unelgee)
niit
```

```python
niit['unelgee_too'] = 1
niit.groupby(['Үнэлгээ'])['unelgee_too'].count()
```

```python
niit_amjilt = niit.loc[niit['Үнэлгээ'] != 'F', 'unelgee_too'].sum()
amjilt = niit_amjilt / 28 * 100
print('Хичээлийн амжилт:', round(amjilt,2), '%')
```

```python
niit_chanar = niit.loc[niit['Үнэлгээ'].isin(['A', 'B','A-','B+','B-']), 'unelgee_too'].sum()
chanar = niit_chanar / 28 * 100
print('Хичээлийн чанар:', round(chanar,2), '%')
```

```python
F_dundaj = niit.loc[niit['Үнэлгээ'] == 'F', 'Нийт'].mean()
F_max = niit.loc[niit['Үнэлгээ'] == 'F', 'Нийт'].max()
F_min = niit.loc[niit['Үнэлгээ'] == 'F', 'Нийт'].min()

print('F mean:', round(F_dundaj,2))
print('F max:' , round(F_max,2))
print('F min:', round(F_min,2))
```
